Import pandas and read a data csv file

```python
import pandas as pd

df = pd.read_csv("orders.csv")
```

Basic pandas filtering

```python
df # shows the dataframe
```

```python
df[df["Category"] == "Electronics"] # displays all the dataframe rows where the category is "Electronics"
```

```python
df[df["Quantity"] > 5] # displays all the dataframe rows where the quantity is greater than 5
```

Multiple filters

```python
df[(df["Category"] == "Electronics") & (df["Country"] == "Japan")] # displays all the dataframe rows where the category is "Electronics" AND the country is "Japan"
```

```python
df[(df["Category"] == "Electronics") | (df["Country"] == "USA")] # displays all the dataframe rows where the category is "Electronics" OR the country is "USA"
```

```python
df[(df["Quantity"] > 2) & (df["Quantity"] <= 10)] # displays all the dataframe rows where the quantity is greater than 2 but less than or equal to 10
```

More advanced filtering

```python
df[df["CustomerName"].str.startswith("D")] # displays all the dataframe rows where the name starts with "D"
```

```python
df[df["CustomerName"].str.endswith("a")] # displays all the dataframe rows where the name ends with "a"
```

```python
df[df["Country"].isin(["USA", "UK", "Canada"])] # displays all the dataframe rows where the country is in the given array
```

```python
asian_client_countries = ["Singapore", "South Korea", "Japan", "India", "Pakistan", "Saudi Arabia"]
df[df["Country"].isin(asian_client_countries)] # displays all the dataframe rows where the country is in the asian_client_countries array
```

```python
df[~df["Country"].isin(asian_client_countries)] # displays all the dataframe rows where the country is NOT in the asian_client_countries array - the tilda (~) reverses the normal operation from "is in" to "is not in"
```

Filtering based on specific values

```python
df.loc[df["CustomerName"] == "Ali Khan"] # displays the row or rows where the "CustomerName" property value is "Ali Khan"
```

```python
df.loc[df["CustomerName"].str.endswith("a")] # displays all the dataframe rows where the name ends with "a"
# NOTE: we did this before using just df[df["CustomerName"].str.endswith("a")] but the loc command is more explicit and adds extra functionality, see below
```

```python
df.loc[df["CustomerName"].str.endswith("a"), ["CustomerName", "Product", "Price"]] # displays all the dataframe rows where the name ends with "a"
# NOTE: now we add a parameter to return selected columns
```
