Import pandas and read a data csv file

```python
import pandas as pd

df = pd.read_csv("orders.csv")
```

Modification using loc

```python
df # shows the dataframe
```

```python
df.loc[df["CustomerName"] == "Anna Ivanova"] # displays the row or rows where the "CustomerName" property value is "Anna Ivanova"
```

```python
annas_product = df.loc[df["CustomerName"] == "Anna Ivanova", "Product"] # finds the row/or rows where the "CustomerName" property value is "Anna Ivanova" and assigns the "Product" property value to annas_product
# <dataframe>.loc[<dataframe>[<value-condition>], <property>] gives the property value/s in that row
annas_product
```

```python
df.loc[df["CustomerName"] == "Anna Ivanova", "Product"] = "Time Machine" # finds the row/or rows where the "CustomerName" property value is "Anna Ivanova" and changes the "Product" property value to "Time Machine"
```

```python
df.loc[df["CustomerName"] == "Anna Ivanova"] # displays the now updated row or rows where the "CustomerName" property value is "Anna Ivanova"
```

Modifying multiple values at once

```python
df.loc[df["Country"] == "USA", "Country"] = "Trumpton" # modifies all countries that are "USA" to "Trumpton
```

```python
df.loc[df["Country"] == "Trumpton"] # shows all the rows where the country is now "Trumpton"
```

Modify property values to uppercase/lowercase

```python
df.loc[df["Country"].str.startswith("U"), "Country"] = df["Country"].str.lower() # modifies all countries beginning with "U" to lowercase
```

```python
df # displays updated dataframe with lowercase countries where they started with a "U"
```

```python
df["Country"] = df["Country"].str.upper() # simplified way that modifies all countries values to uppercase
```

```python
df # shows updated country names in uppercase
```

```python
df.rename(columns={"OrderID": "OrderRef"}, inplace=True) # renames the "OrderID" column to "OrderRef"
```

```python
df # shows updated country names in uppercase
```
